4. Orden de lista de tres numeros (compacto)

```python
import numpy as np

x1 = int(input('ingrese un numero: '))
x2 = int(input('ingrese un numero: '))
x3 = int(input('ingrese un numero: '))
print('\nEl orden de mayor a menor es: ')
X = np.array([x1,x2,x3])


for i in range(len(X)): # cada numero
  i_d = np.where(X == X[i])[0] # indice del numero actual
  X_new = np.delete(X, i_d) # lista sin el numero actual

  if X[i] > X_new[0] and X[i] > X_new[1]: # condicion general

    if X_new[0] < X_new[1]: # condicion restante
      print(f'{X[i]}, {X_new[1]}, {X_new[0]}')
    else:
      print(f'{X[i]}, {X_new[0]}, {X_new[1]}')
```

```python
df = pd.DataFrame( obj_mlc.records( record_set = Record_sets[0].uuid ) )
df.head()
```

```python
# renombrando columnas
df.rename(columns={'life_expectancy.csv/Country':"Country",
                   'life_expectancy.csv/Sum+of+Females++Life+Expectancy':"Females life expectancy",
                   'life_expectancy.csv/Sum+of+Life+Expectancy++(both+sexes)':"Both sex life expectancy",
                   'life_expectancy.csv/Sum+of+Males++Life+Expectancy':"Males life expectancy"}
                   , inplace = True)

df.columns
```

```python
# exprexion para encontrar la "b" y las comillas
pattern= r"b'(.*)'"

# b'     coincidencia con caracteres b'
# (...)  paréntesis se usan para definir caracteres a identificar
#        permite extraer la parte del string que coincide con este grupo
# .      indica que coincide con cualquier carácter, excepto con salto de línea
# *      indica que el carácter anterior (el punto) puede repetirse 0 o más veces
# '      busca una coincidencia literal con el carácter '
```

```python
# Decodificar cadenas-bytes en str, antes de aplicar str.replace
df['Country'] = df['Country'].str.decode('utf-8')

# Aplica RegEx a la columna para limpiarla
df['Country'] = df['Country'].str.replace(pattern, r'\1', regex=True)

# r'\1' se reemplazará el patrón con el texto que se encuentra entre b' y '
```

## 💠 $\ $ **Análisis de Componentes Principales (PCA)**

Es una técnica estadística que se utiliza para la reducción de dimensionalidad de datos complejos. Su objetivo es transformar un conjunto de variables originales, que pueden estar correlacionadas, en un conjunto más pequeño de nuevas variables (componentes principales) que conserven la mayor parte de la información original. Esto simplifica los datos, haciendo más fácil su visualización y análisis, por ejemplo, reduciendo un conjunto de variables de una casa a dos componentes que representen tamaño y ubicación.


* **¿Cómo funciona?**
Transforma los datos: Combina las variables originales en nuevas variables llamadas componentes principales (PC).

* **Crea nuevas variables:** Cada componente principal es una combinación lineal de las variables originales y está ordenado por la cantidad de varianza que explica.

* **Minimiza la pérdida de información:** El PCA busca la dirección en la que los datos varían más (la primera componente principal) y luego las direcciones perpendiculares a esta que expliquen la mayor varianza restante, hasta reducir el número de dimensiones.

* **Resultados:** Los resultados incluyen los eigenvalues (que indican la varianza explicada por cada componente) y los eigenvectors (que definen las direcciones de los nuevos ejes).
Conceptos clave para interpretar los resultados

* **Componentes principales:** Son las nuevas variables no correlacionadas que resumen la información de las originales.

* **Eigenvectors:** Representan la dirección de las nuevas variables (componentes) y se pueden interpretar geométricamente como las direcciones en las que los datos varían más.

* **Eigenvalues:** Muestran la cantidad de varianza explicada por cada componente. Al estar ordenados, permiten saber cuánta información se retiene al reducir la dimensionalidad.

* **Loadings:** Indican el peso o la importancia que tiene cada variable original en cada componente principal, lo que ayuda a entender qué información recoge cada componente.

**Usos**

* **Reducción de la dimensionalidad:** Simplificar conjuntos de datos grandes y complejos para facilitar su análisis.

* **Visualización:** Permite visualizar datos de alta dimensión en un gráfico 2D o 3D, lo cual es imposible de hacer con los datos originales.

* **Preprocesamiento de datos:** Preparar los datos para otras técnicas, como el agrupamiento (clustering) o la regresión, eliminando la redundancia.

```python
# tratamientos de datos
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Graficos
import matplotlib.pyplot as plt
import matplotlib.font_manager
from matplotlib import style

# Preprocesado y modelado
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale
```

### $\ - \ $ **Ajuste del modelo PCA con escalado**

```python
# Ajuste del modelo PCA con escalado de los datos
pca_pipe = make_pipeline(StandardScaler(), PCA())

pca_pipe.fit(datos)

# Extraer el modelo entrenado del pipeline
modelo_pca = pca_pipe.named_steps['pca']
```

```python
# Se convierte a dataframe para anadir nombres a los ejes
pd.DataFrame( data = modelo_pca.components_,
              columns = datos.columns,
              index=['PC1','PC2','PC3','PC4'] )
```

```python
# La correlacion es independiente de la escala, mientaras que la covarianza si depende de la escala de los datos

# Heatmap componentes
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(4,2))
componentes = modelo_pca.components_

plt.imshow(componentes.T, cmap='viridis' , aspect='auto')
plt.yticks(range(len(datos.columns)), datos.columns)
plt.xticks(range(len(datos.columns)), np.arange(modelo_pca.n_components_) +1 )
          # con np arange, se crea un arreglo espacio uniformemente
plt.colorbar();
```

```python
# Con base a la grafica anterior se puede interpretar
# los datos de la siguiente manera:

#1. UrbanPop con PC2 es altamente correlacionada positivo
#2 Rape con PC3 es altamenete correlacionada positivo
#3. Assault con  PC4 es altamente correlacionada positivo
#4. Murder, Assault, Rape, UrbanPop con PC1 es correlacionada positivamente
# 5. El resto con color azul es correlacionada altamente negativa
```

```python
modelo_pca.explained_variance_ratio_
```

```python
# Porcentaje de varianza explicada por cada componente
print('Porcentaje de varianza explicada por cada componente')
print(modelo_pca.explained_variance_ratio_)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,4))
ax.bar( x= np.arange(modelo_pca.n_components_)+1,
        height = modelo_pca.explained_variance_ratio_ )

for x,y in zip (np.arange(len(datos.columns))+1, modelo_pca.explained_variance_ratio_):
  label = round(y,2)
  ax.annotate(label,(x,y), textcoords="offset points",
              xytext=(0,10), ha='center')

  ax.set_xticks(np.arange(modelo_pca.n_components_)+1)
  ax.set_ylim(0,1.1)
  ax.set_title('Porcentaje de varianza explicada por cada componente')
  ax.set_xlabel('Componente principal')
  ax.set_ylabel('Porcentaje varianza explicada')
```

**Conclusion:**

Con base a la grafica de varianza se interpreta que el componente principal 1(PC1) tiene un alto porcentaje de 0.62(%), lo que significa que frente a las otras variables se correlaciona. Precedido del PC2 es de 0.25(%), el pc3 es 0.09(%) y pc4 0.04(%)

n the context of machine learning model validation curves, 'degree' on the horizontal axis represents a specific hyperparameter of the model that is being varied to evaluate its impact on performance.

* For the Random Forest Classifier, degree refers to the number of estimators (n_estimators), which is the number of decision trees in the forest.

The validation curve plots how the model's training and validation scores change as this hyperparameter ('degree') is adjusted.
